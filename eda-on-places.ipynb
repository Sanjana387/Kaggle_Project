# **2.Summary Statistics:**
Summary statistics provide insights into the central tendency, dispersion, and distribution of the numerical columns in the dataset. You can use the describe() function to generate summary statistics for the numerical columns:

```python
# Summary statistics
df.describe().transpose()
```

# **3.Missing Values:**
Missing values can affect the quality and reliability of the data. It is essential to identify and handle missing values appropriately. You can check for missing values in the dataset using the isnull() function and then calculate the total count of missing values in each column using sum():

```python
df.isnull().sum()
```

# **4.Unique Values:**
Understanding the unique values in categorical columns helps us identify the distinct categories and their frequencies. Here's an example of how you can find the unique values and their counts in each column:

```python
\
for column in df.columns:
    if df[column].dtype == object:  # Check for categorical columns
        print("\nUnique values in", column, "column:")
        print( df[column].value_counts())
```

```python

# Sort the DataFrame by 'Crime' column in descending order
df_sorted = df.sort_values(by='Crime', ascending=False)

# Extract the top 10 cities with the highest crime rates
top_10_cities = df_sorted.head(10)

# Create a horizontal bar plot
plt.figure(figsize=(10, 6))
plt.bar(top_10_cities['City'], top_10_cities['Crime'])
plt.xlabel('Crime Rate')
plt.ylabel('City')
plt.title('Top 10 Cities with Highest Crime Rates')
plt.xticks(rotation=90)
plt.show()
```

```python

# Sort the DataFrame by 'Crime' column in descending order
df_sorted = df.sort_values(by='Educ', ascending=False)

# Extract the top 10 cities with the highest crime rates
top_10_cities = df_sorted.head(10)

# Create a horizontal bar plot
plt.figure(figsize=(10, 6))
plt.bar(top_10_cities['City'], top_10_cities['Educ'])
plt.xlabel('Educ Rate')
plt.ylabel('City')
plt.title('Top 10 Cities with Highest Educ Rates')
plt.xticks(rotation=90)
plt.show()
```

```python
# Sort the DataFrame by 'Crime' column in descending order
df_sorted = df.sort_values(by='Transp', ascending=False)

# Extract the city with the highest crime rate (first row)
highest_transp_city = df_sorted.iloc[0]['City']

print("Highest Transpose City:-",highest_transp_city)  
```

```python
plt.boxplot(df['Crime'])
plt.xlabel('Crime')
plt.title('Box Plot of Depth')
plt.show()
```

```python
df_sorted = df.sort_values(by='Pop', ascending=False)

# Extract the city with the highest crime rate (first row)
highest_pop_city = df_sorted.iloc[0]['City']

print("Highest Population City:-",highest_pop_city)  
```

```python
plt.figure(figsize=(8, 6))
df['Crime'].value_counts().plot(kind='line')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of crime Level')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['Econ'].value_counts().plot(kind='line')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of economy Level')
plt.show()
```
